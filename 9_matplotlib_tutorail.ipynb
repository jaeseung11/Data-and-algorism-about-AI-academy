```python
# Lesson: 파이썬 Matplotlib 초급 튜토리얼

# Matplotlib 가져오기
import matplotlib.pyplot as plt

# 선 그래프
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
```

```python
plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')
plt.show()
```

```python
# 산점도
plt.scatter(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.show()
```

```python
# 막대 그래프
x = ['A', 'B', 'C', 'D', 'E']
y = [10, 15, 7, 12, 9]

plt.bar(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Bar Plot')
plt.show()
```

```python
# 히스토그램
data = [1, 2, 2, 3, 3, 3, 4, 4, 5]

plt.hist(data, bins=5)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

```python
# 파이 차트
labels = ['A', 'B', 'C', 'D']
sizes = [20, 30, 25, 15]

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Pie Chart')
plt.show()
```
